## Bilirubin

- It is a red-orange compound that occurs in the normal catabolic pathway that breaks down heme in vertebrates. This catabolism is a necessary process in the body's clearance of waste products that arise from the destruction of aged or abnormal red blood cells #For adults over 18 normal total bilirubin can be up to 1.2 milligrams per deciliter (mg/dl) of blood Direct bilirubin = Conjugated bilirubin + delta bilirubin


## Total Bilirubin

- Total Bilirubin is a combination of direct and indirect bilirubin. <21(micromole/litre), <1.23(milligram/decilitre) direct bilirubin: 1.0-5.1(micromole/litre)0-0.3(milligram/decilitre) An elevation greater than 20 mg/dl suggests severe liver disease. In patient with hepatitis -induced acute liver failure, a serum total bilirubin level >17.5 mg/dL (300 mmol/L)is a criterion for predicting death and the need for liver transplantation.

## Alkaline phosphates

- is secreted by enterocytes and seems to play a pivotal role in intestinal homeostasis and protection as well as in suppressing inflammation via repression of the dowmstream Toll-like receptor (TLR)-4-dependent and my D88-dependent inflammatory cascade The normal range is 44 to 147 international units per liter (IU/L) or 0.73 to 2.45 microkatal per liter (µkat/L) When liver injury occurs, the level of ALP in the circulation increases

## Alamine Aminotransferase:

- on average, the normal levels of SGPT range between 7-56 units/litre of serum. If the SGPT level in your liver blood test are greater than 56 units/litre,it is referred to as the danger level of SGPT and may indicate any form of injury or damage in the liver ALT levels greater than 15 times the normal range indicates severe acute liver cell injury and evaluation should be initiated immidiately Male 8-40 IU/L Female 6-34 IU/L High levels of AST in the blood may be a sign of hepatitis, sirrhosis, mononucleosis or other liver disease


## Total Proteins

- The reference range for total protein is typically 60-80g/L.(It is also sometimes reported as "6.0-8.0g/dl")A high total protein level could indicate dehydration or a certain type of cancer, such as multiple myeloma, that causes protein to accumulate abnormally. If the result of a total protein test is abnormal, further tests will be needed to identify which proteins are too high or too low

## Albumin

- Is a protein made by your liver. Albumin enters your bloodstream and helps keep fluid from leaking out of your blood vessels into other tissues. It is also carries hormones, vitamins and enzymes throuhout your body the normal range is3.4 to 5.4 g/dl934 to 54 g/L).Low level might indicate liver disease

## Albumin and Globulin Ration

- The normal range for albumin/globulin ration is over 1, usually around 1 to 2. That's beacuse there's a bit more albumin than globulin in serum protein. If your body is producing too much or too little of either protein and albumin/globulin (A/G) ratio test meausres the total amount of protein in your blood
